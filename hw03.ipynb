## 引言

### 背景

### 目的
本研究旨在利用公開的 mpg 資料集，分析汽車油耗與汽缸數、車重及產地三者之間的關係。透過視覺化與統計方法，找出影響油耗的主要因素，並比較不同產地車款的油耗表現。

### 待答問題

2. 車重（weight）與油耗（mpg）之間是否存在明顯關聯？
3. 不同產地（origin）的車款，油耗表現有何不同？

2. 針對分析主題（油耗與汽缸數、油耗與車重、油耗與產地）進行資料清理與欄位篩選。

## 結果與分析

### 油耗與汽缸數（cylinders）的關係
我們使用箱型圖來觀察油耗（mpg）與汽缸數（cylinders）之間的關係。

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(x='cylinders', y='mpg', data=tips)
plt.title('不同汽缸數的油耗分布')
plt.xlabel('汽缸數')
plt.ylabel('mpg')
plt.show()
```

**分析結果**：從圖中可以看出，汽缸數越多（如6缸、8缸），油耗表現較差，4缸車款的油耗分布最高且較集中，顯示小排氣量車輛較省油。

### 油耗與車重（weight）的關係
我們觀察油耗（mpg）與車重（weight）的關係，並使用散點圖呈現。


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='weight', y='mpg', data=tips)
plt.title('車重與油耗的關係')
plt.xlabel('車重')
plt.ylabel('mpg')
plt.show()
```

**分析結果**：車重與油耗之間呈現明顯負相關，車重越重的車款，mpg 越低，顯示重量是影響油耗的重要因素。

### 油耗與產地（origin）的比較
我們比較油耗與產地（origin）的關聯，使用箱型圖進行分析。

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='origin', y='mpg', data=tips)
plt.title('不同產地車款的油耗分布')
plt.xlabel('產地')
plt.ylabel('mpg')
plt.show()
```

**分析結果**：
美國車款的油耗普遍較低，歐洲與日本車款油耗表現較佳，顯示不同地區的汽車設計理念與技術差異。

## 結論與建議


### 結論
1. 汽缸數愈多的車款，平均油耗（mpg）明顯較低，顯示大排氣量車輛較不省油；4缸車款的油耗表現最佳。
2. 車重與油耗呈現明顯負相關，車重越重，油耗表現越差，說明車輛自重是影響油耗的重要因素。
3. 日本與歐洲產車款的油耗普遍優於美國車款，反映不同地區汽車設計與技術的差異。
